```python
x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
```

```python
print(type(book))
```

```python
print(book.tag)
```

```python
book_childs = list(book)
print(book_childs)
```

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')

print(name)
print(author)
print(year)
```

```python
x2_str = """
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>
  <book>
    <name>혼자 공부하는 머신러닝+딥러닝</name>
    <author>박해선</author>
    <year>2020</year>
  </book>
</books>
"""
```

```python
books = et.fromstring(x2_str)
print(books.tag)
```

```python
for book in books.findall('book'):
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  print(name)
  print(author)
  print(year)
  print()
```
